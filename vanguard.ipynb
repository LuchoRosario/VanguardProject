COMPLETION  RATE. HYPOTHESIS TESTING

#Two sample t-test para las variables Test y control
#hipotesis nula: no hay diferencia entre las medias de las dos muestras: el total de clientes sujetos al test que completaron el procesos no es mayor que el numero de cliente que tambien completaron pero con el antiguo metodo "Control"
#hipotesis alternativa: hay diferencia entre las medias de las dos muestra: el total de clientes sujetos al test que completaron el procesos es mayor que el numero de cliente que tambien completaron pero con el antiguo metodo "Control"

```python
#Two sample t-test para las variables Test y control
#hipotesis nula: no hay diferencia entre las medias de las dos muestras: el total de clientes sujetos al test que completaron el procesos no es mayor que el numero de cliente que tambien completaron pero con el antiguo metodo "Control"
#hipotesis alternativa: hay diferencia entre las medias de las dos muestra: el total de clientes sujetos al test que completaron el procesos es mayor que el numero de cliente que tambien completaron pero con el antiguo metodo "Control"

from scipy.stats import ttest_ind

test = df_completed[df_completed["Variation"] == "Test"]["completed"]
control = df_completed[df_completed["Variation"] == "Control"]["completed"]

ttest_ind(test, control)

```

```python
#el valor de p es menor que 0.05, por lo que rechazamos la hipotesis nula y aceptamos la hipotesis alternativa. Hay una diferencia significativa entre las medias de las dos muestras. El total de clientes sujetos al test que completaron el proceso es mayor que el numero de cliente que tambien completaron pero con el antiguo metodo "Control"

# EL GRUPO TEST SUPERA EN UN 5% AL GRUPO CONTROL EN CLIENTES QUE COMPLETAN EL PROCESO
```

COMPLETION RATE WITH A COST-EFFECTIVENESS THRESHOLD


```python
# la hypothesis nula es que el competion rate para el Test group es igual o menor que para el Control group incrementado en un 5%
# la hypothesis alternativa es que el competion rate para el Test group es mayor que para el Control group incrementado en un 5%

#realizamos el test teniendo en cuenta esta situacion   
test = df_completed[df_completed["Variation"] == "Test"]["completed"]
control = df_completed[df_completed["Variation"] == "Control"]["completed"]
ttest_ind(test, control + 0.05)
```

```python
#el valor de p es menor que la significancia por lo que descartamos la hipotesis nula y aceptamos la hipotesis alternativa. El competion rate para el Test group es mayor que para el Control group incrementado en un 5%
```

```python
#filtar valores atipicos en time_diff
Q1=df_sin_completos['time_diff'].quantile(0.25)
Q3=df_sin_completos['time_diff'].quantile(0.75)
IQR=Q3-Q1
df_sin_completos=df_sin_completos[(df_sin_completos['time_diff']>(Q1-1.5*IQR)) & (df_sin_completos['time_diff']<(Q3+1.5*IQR))]
```

```python
#quiero hacer un test de hipotesis para ver si hay una diferencia significativa en el tiempo promedio que toma completar cada paso entre los grupos Test y Control
#hipotesis nula: no hay diferencia entre las medias de las dos muestras: el tiempo promedio que toma completar cada paso es similar entre el grupo Test y Control
#hipotesis alternativa: hay diferencia entre las medias de las dos muestra: el tiempo promedio que toma completar cada paso es menor en el grupo Test. 

from scipy.stats import ttest_ind

test = df_sin_completos[df_sin_completos["Variation"] == "Test"]
control = df_sin_completos[df_sin_completos["Variation"] == "Control"]

ttest_ind(test["time_diff"], control["time_diff"])

```

```python
#el valor de p es menor que 0.05, por lo que rechazamos la hipotesis nula.
```

```python
# Crear la columna previous_step para ver el paso anterior del cliente
datafinal['previous_step'] = datafinal.groupby('visit_id')['process_step'].shift(1)
```

```python
datafinal.tail()
```

```python


# Contar los pasos hacia atrás para cada grupo
pasos_hacia_atras_control = datafinal[(datafinal['process_step'] < datafinal['previous_step']) & (datafinal['process_step'] != 4) & (datafinal['Variation'] == 'Control')].shape[0]
pasos_hacia_atras_test = datafinal[(datafinal['process_step'] < datafinal['previous_step']) & (datafinal['process_step'] != 4) & (datafinal['Variation'] == 'Test')].shape[0]

# Calcular el número total de pasos para cada grupo
total_pasos_control = datafinal[datafinal['Variation'] == 'Control']['process_step'].count()
total_pasos_test = datafinal[datafinal['Variation'] == 'Test']['process_step'].count()

# Calcular la tasa de error para cada grupo
tasa_error_control = pasos_hacia_atras_control / total_pasos_control
tasa_error_test = pasos_hacia_atras_test / total_pasos_test

print(f'Tasa de error (Control): {tasa_error_control:.2%}')
print(f'Tasa de error (Test): {tasa_error_test:.2%}')
```
